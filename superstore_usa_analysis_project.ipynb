```python
# find missing values 
df.isnull().sum()
```

```python
# fill the value where value are null 
df["Product Base Margin"]=df["Product Base Margin"].fillna(df["Product Base Margin"].mean(),inplace=True)
```

```python
# check  the value is filled or not 
df.isnull().sum()
```

#### univariate analysis

```python
df["Order Priority"].value_counts()
```

```python
# lets check  why it is showing different Critical:1  and Critical: 1804

df["Order Priority"].unique()
```

```python
# lets replace the critical 
df["Order Priority"]=df["Order Priority"].replace('Critical ','Critical')
```

```python
# lets checkit haas replaced or not 
```

```python
df["Order Priority"].unique()
```

#### graphical representation 

```python
plt.Figure(figsize=(5,4))
sns.countplot(x="Order Priority",data=df)
plt.title("count of Order Priority")
plt.savefig("count of Order Priority.jpg")
plt.show
```

#### shipping mode 

```python
df["Ship Mode"].value_counts()
```

```python
#  unique values in ship mode column 
df["Ship Mode"].value_counts().index
```

```python
# for piechart lets put above value in x 
x=df["Ship Mode"].value_counts().index
y=df["Ship Mode"].value_counts().values
```

```python
plt.figure(figsize=(5,4))

plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.legend(loc=2    )
plt.show()
```
